```python
import folium
from folium.plugins import HeatMap
import json


map = folium.Map(location=[47.36635101042978, 8.541361139590903], tiles="Cartodb Positron", zoom_start=13)

# Load GeoJSON data
with open("/content/RoadTrafficAccident_ZH_city_epsg4326.geojson") as f:
    geojson_data = json.load(f)

# Extract data points for HeatMap
heat_data = []
for feature in geojson_data['features']:
    if feature['geometry']['type'] == 'Point' and feature['properties']['AccidentSeverityCategory'] == 'as3' and feature['properties']['AccidentInvolvingBicycle'] == 'true':
        coordinates = feature['geometry']['coordinates']
        heat_data.append([coordinates[1], coordinates[0]])  # Latitude, Longitude

# Create and add HeatMap
HeatMap(heat_data).add_to(map)

map
```
